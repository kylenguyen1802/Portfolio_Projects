```python
# Import libraries
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #Adjust the configuration of the plots


# Read the data
df = pd.read_csv(r'C:\Users\kylen\Downloads\ATA_Projects\Python_Movie_Project\movies.csv')
```

```python
# Look at the data
df.head()
```

```python
# Look for any missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Drop any row with missing values
df = df.dropna()
```

```python
# Data types for our columns
df.dtypes
```

```python
# Change the data types for budget and gross columns.
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
# Fix the year to match with the year in the released column
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
df
```

```python
# Sort by movies with highest-lowest gross
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Look for and drop duplicates
df.drop_duplicates()
```

```python
# Budget high correlation (guess)
# Company high correlation (guess)
```

```python
# Scatter plot with budget vs gross
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget')
plt.show()
```

```python
df.head()
```

```python
# See if the two variables are correlated. Plot budget vs gross using seaborn
sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
plt.show()
```

```python
# Start looking at correlation
```

```python
# correlation methods: pearson(default), kendall, spearman
df_numeric = df.select_dtypes(include='number')
df_numeric.corr(method='pearson')
```

```python
# High correlation between budget and gross (correct)
```

```python
correlation_matrix = df_numeric.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```
